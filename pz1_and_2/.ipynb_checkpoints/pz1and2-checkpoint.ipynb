# Отчёт: Модели машинного обучения на TensorFlow и Keras

**Цель:** загрузить данные, построить и обучить модели с учителем и без учителя, оценить их качество.

**Датасет:** Fashion-MNIST (изображения одежды и обуви, 10 классов, 28×28 пикселей).

## 1. Импорт библиотек и загрузка данных

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()

class_names = ['T-shirt', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

print('Обучающая выборка:', x_train.shape, y_train.shape)
print('Тестовая выборка:', x_test.shape, y_test.shape)
print('Диапазон пикселей:', x_train.min(), '-', x_train.max())
```

## 2. Предобработка данных

```python
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
x_train = np.expand_dims(x_train, axis=-1)
x_test = np.expand_dims(x_test, axis=-1)

x_train_flat = x_train.reshape(-1, 784)
x_test_flat = x_test.reshape(-1, 784)

print('Для классификатора:', x_train.shape)
print('Для автоэнкодера:', x_train_flat.shape)
```

```python
plt.figure(figsize=(10, 6))
for i in range(15):
    plt.subplot(3, 5, i + 1)
    plt.imshow(x_train[i].squeeze(), cmap='gray')
    plt.title(class_names[y_train[i]])
    plt.axis('off')
plt.tight_layout()
plt.show()
```

## 3. Модель с учителем: свёрточная сеть для классификации

```python
model_supervised = keras.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)), # первый слой - извлечение простых признаков
    layers.MaxPooling2D((2, 2)), # уменьшает размер изображения для уменьшения кол во вычислений
    layers.Conv2D(64, (3, 3), activation='relu'), # 64 фильтра - больше признаков
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(), # преобразование в вектор
    layers.Dense(64, activation='relu'), # комбинирование признаков
    layers.Dropout(0.5), # борьба с переобучением 
    layers.Dense(10, activation='softmax') # выходной слой (10 признаков)
])

model_supervised.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model_supervised.summary()
```

```python
history_sup = model_supervised.fit(
    x_train, y_train,
    epochs=10,
    batch_size=128,
    validation_split=0.1,
    verbose=1
)
```

### Оценка качества модели с учителем

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history_sup.history['accuracy'], label='train')
plt.plot(history_sup.history['val_accuracy'], label='val')
plt.title('Точность')
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(history_sup.history['loss'], label='train')
plt.plot(history_sup.history['val_loss'], label='val')
plt.title('Функция потерь')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
loss_sup, acc_sup = model_supervised.evaluate(x_test, y_test, verbose=0)
print(f'Тестовая точность (accuracy): {acc_sup:.4f}')
print(f'Тестовая функция потерь: {loss_sup:.4f}')

y_pred = np.argmax(model_supervised.predict(x_test), axis=1)
print('\nОтчёт по классам:')
print(classification_report(y_test, y_pred, target_names=class_names))
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_names, yticklabels=class_names, cmap='Blues')
plt.title('Матрица ошибок (модель с учителем)')
plt.ylabel('Истинный класс')
plt.xlabel('Предсказанный класс')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## 4. Модель без учителя: автоэнкодер

```python
encoding_dim = 32

encoder = keras.Sequential([
    layers.Dense(256, activation='relu', input_shape=(784,)),
    layers.Dense(128, activation='relu'),
    layers.Dense(encoding_dim, activation='relu')
])

decoder = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(encoding_dim,)),
    layers.Dense(256, activation='relu'),
    layers.Dense(784, activation='sigmoid')
])

autoencoder = keras.Sequential([encoder, decoder])
autoencoder.compile(optimizer='adam', loss='mse')
autoencoder.summary()
```

```python
history_ae = autoencoder.fit(
    x_train_flat, x_train_flat,
    epochs=15,
    batch_size=256,
    validation_split=0.1,
    verbose=1
)
```

### Оценка качества автоэнкодера

```python
plt.figure(figsize=(8, 4))
plt.plot(history_ae.history['loss'], label='train')
plt.plot(history_ae.history['val_loss'], label='val')
plt.title('Функция потерь (MSE) автоэнкодера')
plt.legend()
plt.xlabel('Эпоха')
plt.show()
```

```python
mse_test = autoencoder.evaluate(x_test_flat, x_test_flat, verbose=0)
print(f'Средняя квадратичная ошибка реконструкции на тесте: {mse_test:.6f}')

# Визуализация реконструкций
decoded = autoencoder.predict(x_test_flat[:10])
plt.figure(figsize=(14, 4))
for i in range(5):
    plt.subplot(2, 5, i + 1)
    plt.imshow(x_test[i].squeeze(), cmap='gray')
    plt.title('Исходное')
    plt.axis('off')
    plt.subplot(2, 5, i + 6)
    plt.imshow(decoded[i].reshape(28, 28), cmap='gray')
    plt.title('Реконструкция')
    plt.axis('off')
plt.suptitle('Автоэнкодер: исходные и восстановленные изображения')
plt.tight_layout()
plt.show()
```

## 5. Выводы

```python
from sklearn.metrics import precision_score, recall_score, f1_score


precision_sup = precision_score(y_test, y_pred, average='weighted')
recall_sup = recall_score(y_test, y_pred, average='weighted')
f1_sup = f1_score(y_test, y_pred, average='weighted')

print('=' * 60)
print('МОДЕЛЬ С УЧИТЕЛЕМ (CNN-классификатор)')
print('=' * 60)
print(f'  Accuracy (точность):           {acc_sup:.4f} ({acc_sup*100:.2f}%)')
print(f'  Loss (функция потерь):         {loss_sup:.4f}')
print(f'  Precision (точность по классам): {precision_sup:.4f}')
print(f'  Recall (полнота):              {recall_sup:.4f}')
print(f'  F1-score (взвешенное):        {f1_sup:.4f}')
print()
print('=' * 60)
print('МОДЕЛЬ БЕЗ УЧИТЕЛЯ (автоэнкодер)')
print('=' * 60)
print(f'  MSE реконструкции на тесте:    {mse_test:.6f}')
print(f'  Размерность латентного пространства: {encoding_dim}')
print()
print('=' * 60)
```

### 5.2. Качественный анализ и выводы

**Модель с учителем (свёрточная нейросеть):**

- Архитектура: три свёрточных блока (32 64 64 фильтра), полносвязный слой с Dropout, выход softmax на 10 классов.
- По графикам обучения видно сходимость: train и val accuracy растут, loss уменьшается. Разрыв между train и val указывает на степень переобучения.
- Точность на тесте отражает долю правильно классифицированных изображений. Precision и Recall по классам показывают, какие категории (T-shirt, Shirt, Coat и т.д.) распознаются лучше, а какие чаще путаются между собой.
- Матрица ошибок наглядно показывает типичные путаницы (например, Shirt  T-shirt, Pullover  Coat).

**Модель без учителя (автоэнкодер):**

- Архитектура: энкодер 784 256 128 32, декодер 32 128 256 784. Сжатие в 32-мерное латентное пространство.
- Метрика качества — MSE реконструкции: чем ниже, тем точнее восстанавливаются изображения из сжатого представления.
- Визуальная оценка реконструкций показывает, насколько сохраняются контуры и детали при сжатии. Типично теряются мелкие детали, но общая форма остаётся узнаваемой.

**Сравнение подходов:**

| Критерий | С учителем (CNN) | Без учителя (автоэнкодер) |
|----------|------------------|---------------------------|
| Задача | Классификация по меткам | Реконструкция, сжатие |
| Метки | Используются | Не используются |
| Основная метрика | Accuracy, F1 | MSE реконструкции |
| Применение | Распознавание классов | Снижение размерности, шумоподавление, визуализация |

Обе модели успешно обучаются на Fashion-MNIST. CNN решает задачу классификации с высокой точностью; автоэнкодер демонстрирует способность сжимать изображения в компактное представление с приемлемым качеством восстановления.
